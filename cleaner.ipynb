```python
# Standardize country names to ISO codes
country_codes = {
    'USA': 'US',
    'United States': 'US',
    'United Kingdom': 'UK',
    'Germany': 'GM',
    # Add more country codes as needed
}

df['country_code'] = df['country'].map(country_codes)
```

```python
import re

def clean_phone_number(phone_number):
    # Remove all non-numeric characters
    phone_number = re.sub('[^0-9]', '', phone_number)
    
    # Check the country code and replace with the local dialing code
    if phone_number.startswith('44') or phone_number.startswith('0044'):
        phone_number = '0' + phone_number[2:]
    elif phone_number.startswith('1') or phone_number.startswith('001'):
        phone_number = '1' + phone_number[1:]
    elif phone_number.startswith('49') or phone_number.startswith('0049'):
        phone_number = '0' + phone_number[2:]
    
    return phone_number
```

```python
df['phone_number'] = df['phone_number'].apply(clean_phone_number)
```

```python
# define conversion factors for different units
unit_to_kg = {
    'kg': 1.0, # kg to kg
    'g': 0.001, # grams to kg
    'mg': 0.000001, # milligrams to kg
    'ml': 0.001 # milliliters to kg
}
```

```python
def convert_to_kg(value):
    if pd.isnull(value):
        return value
    else:
        try:
            num, unit = re.match(r'^(\d*\.?\d+)\s*([a-zA-Z]+)$', value).groups()
        except AttributeError:
            return value
        unit = unit.lower() # convert unit to lowercase
        if unit not in unit_to_kg:
            return value
        else:
            kg_value = float(num) * unit_to_kg[unit]
            return kg_value
```

```python

# apply the conversion function to the weight column
json_data['weight'] = json_data['weight'].apply(convert_to_kg)
```
